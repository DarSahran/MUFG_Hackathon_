```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from tqdm import tqdm
from scipy.stats import chi2_contingency

# ---------- Separate categorical columns ----------
categorical_df = data.select_dtypes(include=['object', 'category'])

# ---------- Function to calculate Cramer's V ----------
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

# ---------- Create Cramer's V matrix ----------
def cramers_v_matrix(df):
    cols = df.columns
    n = len(cols)
    matrix = pd.DataFrame(np.zeros((n, n)), index=cols, columns=cols)
    
    for col1 in tqdm(cols, desc="Calculating Cramer's V"):
        for col2 in cols:
            if col1 == col2:
                matrix.loc[col1, col2] = 1.0
            else:
                matrix.loc[col1, col2] = cramers_v(df[col1], df[col2])
    return matrix

# ---------- Generate the heatmap ----------
cv_matrix = cramers_v_matrix(categorical_df)

plt.figure(figsize=(40, 40))
sns.heatmap(cv_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Cramer's V Correlation Heatmap (Categorical Variables)")
plt.xticks(fontsize=12, rotation=45, ha='right')
plt.yticks(fontsize=12, rotation=0)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import mutual_info_classif
import seaborn as sns
import matplotlib.pyplot as plt

# Function to compute Mutual Information between all categorical columns
def compute_mutual_info(df):
    cols = df.columns
    mi_matrix = pd.DataFrame(index=cols, columns=cols, dtype=float)

    for col1 in cols:
        for col2 in cols:
            if col1 == col2:
                mi_matrix.loc[col1, col2] = 1.0  # perfect association with itself
            else:
                # Encode categories as integers
                encoded_df = df[[col1, col2]].apply(lambda x: pd.factorize(x)[0])
                x = encoded_df[col1]
                y = encoded_df[col2]
                # mutual_info_classif expects 2D array for X
                mi = mutual_info_classif(x.to_frame(), y, discrete_features=True, random_state=42)
                mi_matrix.loc[col1, col2] = mi[0]
    return mi_matrix.astype(float)

# Calculate MI for categorical_df
mi_df = compute_mutual_info(categorical_df)

# Plot heatmap
plt.figure(figsize=(40, 40))
sns.heatmap(mi_df, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Mutual Information Heatmap (Categorical Variables)")
plt.show()
```
